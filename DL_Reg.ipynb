```python
#boston dataset
dataset = pd.read_csv("/content/train.csv")
dataset.head(5)
```

```python
#missing values
dataset.isnull().sum()
```

```python
#check the data type
dataset.info()
```

```python
#split the data into dependent vble and independent vble
x= dataset.iloc[:,1:-1]
y = dataset.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
#split the data into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=101)
print(x_train.shape, x_test.shape, y_train.shape,y_test.shape)
```

```python
#feature scaling complusory for DL
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# building multilayer perceptron model/ dnn
from tensorflow import keras
from keras.layers import Dense
from keras import Sequential
```

```python
model = Sequential()
model.add(Dense(6,activation = 'relu', input_dim = 13))#hidden layer 1

#model.add(Dense(6,activation = 'relu'))
model.add(Dense(6,activation = 'relu'))#hidden layer 2
model.add(Dense(1,activation = 'linear'))#output
model.compile(optimizer = 'sgd', loss = 'mse')
```

```python
model.summary()
```

```python
model.fit(x_train, y_train, epochs= 100, validation_data=(x_test,y_test))
```
